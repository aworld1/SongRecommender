```python
import gzip
import math
import random
import preprocess_tools
import numpy
from collections import defaultdict
```

```python

jsonRange = []
for i in range(0,48):
    jsonRange.append(i)
l = ["mpd.slice." + str(i * 1000) + "-" + str((i + 1) * 1000 - 1) + ".json" for i in jsonRange]
data = preprocess_tools.dataloader_pipeline(l)
```

```python
def shuffleData(data):
    shuffledData = numpy.array(data)
    numpy.random.shuffle(shuffledData)
    return shuffledData

tracks_per_user, users_per_track, users_per_artist = preprocess_tools.build_relevant_ds(data)
```

```python
sData = shuffleData(data)
dataTrain = sData[:39000]
dataTest = sData[39000:]
```

```python
def Jaccard(s1, s2):
    numer = len(s1.intersection(s2))
    denom = len(s1.union(s2))
    if denom == 0:
        return 0
    return numer / denom
```

```python
def mostSimilar(i, N):
    similarities = []
    users = users_per_track[i]
    for i2 in users_per_track:
        if i2 == i: continue
        sim = Jaccard(set(users), set(users_per_track[i2]))
        similarities.append((float(sim),str(i2)))
    similarities.sort(reverse=True)
    return similarities[:N]
```

```python
trackNames = list(users_per_track.keys())

def getUnpickedTrack(u):
    while True:
        r = random.randint(0, len(trackNames))
        if trackNames[r] not in tracks_per_user[u]:
            return trackNames[r]
```

```python
correct = 0
total = 0
thresh = 0.01
for pl in dataTest[:100]:
    tracks = pl["tracks"]
    u = pl["pid"]
    if len(tracks) <= 1:
        continue
    correctTrack = ""
    if len(tracks_per_user[u]) > 0:
        correctTrack = tracks_per_user[u][-1]
    nonTrack = getUnpickedTrack(u)
    simC = [0]
    simN = [0]
    usersPerTrackWoCorrect = [user for user in users_per_track[correctTrack] if user is not u]
    for t in tracks_per_user[u][:-1]:
        usersPerTrackWoCurrent = [user for user in users_per_track[t] if user is not u]
        simC.append(Jaccard(set(usersPerTrackWoCurrent), set(usersPerTrackWoCorrect)))
        simN.append(Jaccard(set(usersPerTrackWoCurrent), set(users_per_track[nonTrack])))
    if max(simC) >= thresh:
        correct += 1
    if max(simN) < thresh:
        correct += 1
    print(len(tracks_per_user[u]))
    total += 2
print("Correct: " + str(correct) + "\nTotal: " + str(total))
acc = correct / total
print("Accuracy: " + str(acc*100) + "%")
```
